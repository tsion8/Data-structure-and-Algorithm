**1: Dynamic Array / Array-Based list**

```python
# Dynamic Array Implementation (Without Using Python List Methods for Main Logic)
class DynamicArray:
    def __init__(self):
        self.capacity = 2
        self.count = 0
        self.array = [None] * self.capacity

    def resize(self):
        self.capacity = self.capacity * 2
        new_array = [None] * self.capacity
        for i in range(self.count):
            new_array[i] = self.array[i]
        self.array = new_array

    def append(self, value):
        if self.count == self.capacity:
            self.resize()
        self.array[self.count] = value
        self.count += 1

    def get(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")   # ← fixed
        return self.array[index]

    def set(self, index, value):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")   # ← fixed
        self.array[index] = value

    def insert(self, index, value):
        if index < 0 or index > self.count:
            raise IndexError("Index out of range")   # ← fixed
        if self.count == self.capacity:
            self.resize()
        for i in range(self.count, index, -1):
            self.array[i] = self.array[i - 1]
        self.array[index] = value
        self.count += 1

    def delete(self, index):
        if index < 0 or index >= self.count:
            raise IndexError("Index out of range")   # ← fixed
        for i in range(index, self.count - 1):
            self.array[i] = self.array[i + 1]
        self.array[self.count - 1] = None
        self.count -= 1

    def size(self):
        return self.count

    def display(self):
        for i in range(self.count):
            print(self.array[i], end=" ")
        print()

# ---------------------------------------------------
# Bioinformatics Application
# ---------------------------------------------------
dna = DynamicArray()
sequence = "ATGCAGT"
for nucleotide in sequence:
    dna.append(nucleotide)

print("Original DNA Sequence:")
dna.display()

print("\nNucleotide at position 3:")
print(dna.get(2))

dna.insert(3, 'T')
print("\nDNA Sequence after inserting mutation at position 4:")
dna.display()

dna.delete(1)
print("\nDNA Sequence after deleting nucleotide at position 2:")
dna.display()

dna.set(4, 'C')
print("\nDNA Sequence after replacement:")
dna.display()

print("\nFinal Size of DNA Sequence:")
print(dna.size())
```

**Report**

A Dynamic Array was implemented using a Python class called DynamicArray. The data structure automatically increases its capacity when the array becomes full. Several operations were implemented, including append(), get(), set(), insert(), delete(), and size(). The append(), get(), and set() operations have a time complexity of O(1) on average, while insert() and delete() have a time complexity of O(n) because elements must be shifted. The resize operation also has a time complexity of O(n) since all elements are copied into a new larger array.
In the bioinformatics application, the dynamic array was used to store a DNA sequence such as A, T, G, C, A, G, and T. Different operations were performed on the sequence, including accessing a nucleotide at a specific position, inserting a mutation, deleting a nucleotide, and replacing one nucleotide with another. This demonstrates how dynamic arrays can be used in bioinformatics for DNA sequence editing, mutation analysis, and genome data processing.

**2: Single Linked List**

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def prepend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_after(self, target, data):
        current = self.head
        while current and current.data != target:
            current = current.next
        if current:
            new_node = Node(data)
            new_node.next = current.next
            current.next = new_node

    def delete(self, data):
        if not self.head:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        current = self.head
        while current.next and current.next.data != data:
            current = current.next
        if current.next:
            current.next = current.next.next

    def search(self, data):
        current = self.head
        while current:
            if current.data == data:
                return True
            current = current.next
        return False

    def display(self):
        nodes = []
        current = self.head
        while current:
            nodes.append(current.data)
            current = current.next
        return " → ".join(nodes)


# Bioinformatics Application
assembly = SinglyLinkedList()
assembly.append("Contig1")
assembly.append("Contig2")
assembly.append("Contig3")

print("Original assembly:", assembly.display())

# Add a new contig
assembly.append("Contig4")
print("After adding Contig4:", assembly.display())

# Remove a low-quality contig
assembly.delete("Contig2")
print("After removing Contig2:", assembly.display())

# Search for a specific contig
print("Search Contig3:", assembly.search("Contig3"))

# Display the assembly path
print("Final assembly path:", assembly.display())
```

**Report**
  
A Singly Linked List was implemented using two classes: a Node class and a SinglyLinkedList class. Each node stores data and a reference to the next node in the list. The linked list supports several operations including append(), prepend(), insert_after(), delete(), search(), and display(). The prepend() operation has a time complexity of O(1) because insertion occurs at the beginning of the list, while append(), insert_after(), delete(), search(), and display() have a time complexity of O(n) because the list may need to be traversed node by node.
In the bioinformatics application, the singly linked list was used to represent contigs in a simple genome assembly path such as Contig1 → Contig2 → Contig3. The program performed operations including adding a new contig, removing a low-quality contig, searching for a specific contig, and displaying the assembly path. This demonstrates how linked lists can be used in bioinformatics to model genome assembly processes where DNA fragments are connected sequentially.


**3: Stack**

```python
# Stack Implementation
class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):
        self.items += [item]

    def pop(self):
        if self.is_empty():
            raise IndexError("Stack is empty")   # ← fixed
        item = self.items[-1]
        del self.items[-1]                       # ← fixed
        return item

    def peek(self):
        if self.is_empty():
            raise IndexError("Stack is empty")   # ← fixed
        return self.items[-1]

    def is_empty(self):
        return len(self.items) == 0

    def size(self):
        return len(self.items)

    def display(self):
        print(self.items)

# -----------------------------------
# Bioinformatics Application
# Sequence Alignment Traceback
# -----------------------------------
alignment_stack = Stack()

alignment_stack.push("Match")
alignment_stack.push("Match")
alignment_stack.push("Insertion")
alignment_stack.push("Match")
alignment_stack.push("Deletion")

print("Alignment Operations in Stack:")
alignment_stack.display()

print("\nTop Operation:")
print(alignment_stack.peek())

print("\nTraceback Path (Reverse Order):")
while not alignment_stack.is_empty():
    print(alignment_stack.pop())

print("\nFinal Stack Size:")
print(alignment_stack.size())
```

**Report**

A Stack data structure was implemented using a Python class. The stack follows the LIFO (Last In First Out) principle, where the last element added is the first element removed. The implemented operations include push(), pop(), peek(), is_empty(), and size(). All of these operations have a time complexity of O(1), which makes stack operations efficient because elements are added and removed only from the top of the stack.
In the bioinformatics application, the stack was used to simulate sequence alignment traceback. Alignment operations such as Match, Match, Insertion, Match, and Deletion were pushed onto the stack and later popped in reverse order to reconstruct the alignment path. This demonstrates how stacks are useful in bioinformatics algorithms such as Needleman-Wunsch and Smith-Waterman sequence alignment, where traceback operations are performed from the end of the alignment back to the beginning.

**4: Queue**

```python
# Queue Implementation

class Queue:
    def __init__(self):
        self.items = []

    # Add item to queue
    def enqueue(self, item):
        self.items += [item]

    # Remove first item
    def dequeue(self):
        if self.is_empty():
            return "Queue is empty"

        front_item = self.items[0]
        self.items = self.items[1:]
        return front_item

    # View first item
    def front(self):
        if self.is_empty():
            return "Queue is empty"

        return self.items[0]

    # Check if queue is empty
    def is_empty(self):
        return len(self.items) == 0

    # Return queue size
    def size(self):
        return len(self.items)

    # Display queue
    def display(self):
        print(self.items)


# -----------------------------------
# Bioinformatics Application
# Sequencing Read Processing
# -----------------------------------

reads_queue = Queue()

# Enqueue sequencing reads
reads_queue.enqueue("Read1")
reads_queue.enqueue("Read2")
reads_queue.enqueue("Read3")
reads_queue.enqueue("Read4")

print("Sequencing Reads in Queue:")
reads_queue.display()

# Show first read
print("\nFirst Read:")
print(reads_queue.front())

# Process reads one by one
print("\nProcessing Reads:")

while not reads_queue.is_empty():
    print(reads_queue.dequeue())

# Final queue size
print("\nFinal Queue Size:")
print(reads_queue.size())
```

**Report**
  
A Queue data structure was implemented using a Python class. The queue follows the FIFO (First In First Out) principle, meaning the first element added is the first one removed. The implemented operations include enqueue(), dequeue(), front(), is_empty(), and size(). The enqueue(), front(), is_empty(), and size() operations have a time complexity of O(1). However, dequeue() has a time complexity of O(n) in this implementation because elements must be shifted when removing the front element.
In the bioinformatics application, the queue was used to simulate the processing of sequencing reads in the order they arrive. Reads such as Read1, Read2, Read3, and Read4 were enqueued into the queue and then processed one by one using dequeue(). This demonstrates how queues are used in bioinformatics pipelines to ensure sequencing reads are processed in the correct arrival order during data analysis.
