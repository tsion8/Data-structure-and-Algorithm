```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# Build tree
root = Node("A")
root.left = Node("B")
root.right = Node("C")

root.left.left = Node("D")
root.left.right = Node("E")
root.right.right = Node("F")


# -----------------------
# PRE-ORDER TRAVERSAL
# Root → Left → Right
# Output: A B D E C F
# -----------------------
def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)


# -----------------------
# IN-ORDER TRAVERSAL
# Left → Root → Right
# Output: D B E A C F
# -----------------------
def inorder(node):
    if node:
        inorder(node.left)
        print(node.data, end=" ")
        inorder(node.right)


# -----------------------
# POST-ORDER TRAVERSAL
# Left → Right → Root
# Output: D E B F C A
# -----------------------
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")


# -----------------------
# LEVEL-ORDER TRAVERSAL
# BFS (Level by level)
# Output: A B C D E F
# -----------------------
def levelorder(root):
    if root is None:
        return

    q = deque([root])

    while q:
        node = q.popleft()
        print(node.data, end=" ")

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)


# -----------------------
# RUN ALL FUNCTIONS
# -----------------------
print("Pre-order:")
preorder(root)
print("\n")

print("In-order:")
inorder(root)
print("\n")

print("Post-order:")
postorder(root)
print("\n")

print("Level-order:")
levelorder(root)
print()
```

Search for:

        50
       /  \
     30    70
    / \    / \
   20 40  60 80

```python
class BSTNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


def insert(root, key):
    if root is None:
        return BSTNode(key)

    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)

    return root


def search(root, key):
    if root is None:
        return False

    if root.key == key:
        return True
    elif key < root.key:
        return search(root.left, key)
    else:
        return search(root.right, key)


values = [50, 30, 70, 20, 40, 60, 80]

root = None
for v in values:
    root = insert(root, v)

print(search(root, 40))  # True
print(search(root, 90))  # False
```

```python
class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


def insert(root, key):
    if root is None:
        return Node(key)

    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)

    return root


# Simple sideways tree printer
def print_tree(node, level=0):
    if node is not None:
        print_tree(node.right, level + 1)
        print("    " * level + str(node.key))
        print_tree(node.left, level + 1)


# Build tree
values = [10, 20, 30, 40, 50]
root = None

for v in values:
    root = insert(root, v)


# Print tree
print_tree(root)
```

```python
class BSTNode:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None


# Insert function
def insert(root, key):
    if root is None:
        return BSTNode(key)

    if key < root.key:
        root.left = insert(root.left, key)
    else:
        root.right = insert(root.right, key)

    return root


# Height function
def height(root):
    if root is None:
        return -1   # height in terms of edges

    left_height = height(root.left)
    right_height = height(root.right)

    return 1 + max(left_height, right_height)


# Build BST
values = [10, 20, 30, 40, 50]

root = None
for v in values:
    root = insert(root, v)

# Output height
print("Height of tree:", height(root))
```

## ***Part VI: Bioinformatics Application***

           Ancestor
        /      \
   Mammals     Birds
    /   \
 Human   Cow

```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# -----------------------
# Build the tree
#        Ancestor
#        /      \
#   Mammals     Birds
#    /   \
# Human   Cow
# -----------------------

root = Node("Ancestor")
root.left = Node("Mammals")
root.right = Node("Birds")

root.left.left = Node("Human")
root.left.right = Node("Cow")


# -----------------------
# Traversal functions
# -----------------------

# Pre-order (Root → Left → Right)
def preorder(node):
    if node:
        print(node.data, end=" ")
        preorder(node.left)
        preorder(node.right)


# Post-order (Left → Right → Root)
def postorder(node):
    if node:
        postorder(node.left)
        postorder(node.right)
        print(node.data, end=" ")


# Level-order (BFS)
def levelorder(root):
    if root is None:
        return

    q = deque([root])

    while q:
        node = q.popleft()
        print(node.data, end=" ")

        if node.left:
            q.append(node.left)
        if node.right:
            q.append(node.right)


# -----------------------
# Height of tree
# -----------------------
def height(node):
    if node is None:
        return -1   # height in edges

    return 1 + max(height(node.left), height(node.right))


# -----------------------
# Run outputs
# -----------------------

print("Pre-order:")
preorder(root)
print()

print("Post-order:")
postorder(root)
print()

print("Level-order:")
levelorder(root)
print()

print("Height:", height(root))


# -----------------------
# Root, internal, leaf nodes (manual classification)
# -----------------------
print("\nRoot node: Ancestor")
print("Internal nodes: Ancestor, Mammals")
print("Leaf nodes: Human, Cow, Birds")
```
